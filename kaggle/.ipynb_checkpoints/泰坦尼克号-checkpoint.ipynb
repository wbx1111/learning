```python
from sklearn.model_selection import StratifiedKFold
```

```python
titanic.columns
```

```python
features = np.array(['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked'])
data = titanic[features].values.copy()
target = titanic.Survived.values.copy()
```

```python
skf = StratifiedKFold()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
accs = []
for train, test in skf.split(data, target):
    # 取数据
    X_train = data[train]
    y_train = target[train]
    X_test = data[test]
    y_test = target[test]
    
    linear = LinearRegression()
    linear.fit(X_train, y_train)
    # 把线性回归预测得到的数字当成生还的可能性. 数字越大, 越可能生还, 数字越小生还可能性越低. 
    y_ = linear.predict(X_test)
#     print(y_)
    # 设定阈值, 把生还可能性转化成具体的类别. >= 0.5 预测为生还. 
    y_pred = (y_ >= 0.5)
    # 计算准确率
    acc = (y_pred == y_test).mean()
    print(acc)
    accs.append(acc)
    
print('平均准确率: ', np.array(accs).mean())
```

```python
# 封装一下画特征重要性的函数
def draw_feature_importances(x, importances):
    plt.bar(np.arange(x), np.sort(importances)[::-1])
    _ = plt.xticks(np.arange(x), features[np.argsort(importances)[::-1]])
```

```python
# 想办法取title
Braund, Mr. Owen Harris
Heikkinen, Miss. Laina
Futrelle, Mrs. Jacques Heath (Lily May Peel)
```

```python
# 从有规律的字符串中提取字符串, 用正则. 
name = 'Futrelle, Mrs. Jacques Heath (Lily May Peel)'
```

```python
import re
```

```python
# 默认是贪婪匹配
re.match(r'.*?,\s(.*?)\.\s.*', name).group(1)
```

```python
# 头衔在一定程度上能反映一个人的身份. 
titanic['Title'] = titanic['Name'].map(lambda name: re.match(r'.*?,\s(.*?)\.\s.*', name).group(1))
```

```python
# col表示上校, rev表示牧师, Mlle表示法国小姐, Major陆军少校, Sir爵士, Capt上尉, Countess女伯爵, Jonkheer无名贵族, Don阁下,尊称,Mme夫人 Dona葡萄牙语的夫人小姐
# 尽量把相同的意思的title归成一类.
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Dr": 5, "Rev": 6, "Major": 7, "Col": 7, "Mlle": 8, "Mme": 8, "Don": 9, "Lady": 10, "the Countess": 10, "Jonkheer": 10, "Sir": 9, "Capt": 7, "Ms": 2}
```

```python
titanic['Title'].unique()
```

```python
titanic['Title'] = titanic['Title'].map(title_mapping)
```

```python
# 剩下的继续做特征重要性, 继续选择特征, 进行训练, 观察结果. 最终找到结果最好的算法. 
```

```python
# stacking, 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_new, target, test_size=0.2)
```

```python
logistic = LogisticRegression(max_iter=1000)
logistic.fit(X_train, y_train)
```

```python
y_logistic = logistic.predict(X_test)
```

```python
# 有好几种stacking的做法. 直接对结果进行投票
```

```python
# 第二种直接拿概率, 概率进行加权平均. 
y_proba_logistic = logistic.predict_proba(X_test)
```

```python
svc = SVC(kernel='rbf', probability=True)
svc.fit(X_train, y_train)
y_proba_svc = svc.predict_proba(X_test)
```

```python
y_proba = 0.8 * y_proba_logistic[:, 1] + 0.2 * y_proba_svc[:, 1]
```

```python
(y_test == (y_proba >= 0.5)).mean()
```
