```python
features = np.array(['policy_state', 'policy_csl', 'insured_sex',
       'insured_education_level', 'insured_occupation', 'insured_hobbies',
       'insured_relationship', 'incident_type',
       'collision_type', 'incident_severity', 'authorities_contacted',
       'incident_state', 'incident_city', 'incident_location',
       'property_damage', 'police_report_available', 'auto_make',
       'auto_model'])
```

```python
data = df[features].values.copy()
target = df.fraud_reported.values.copy()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data_scaler = scaler.fit_transform(data)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_scaler, target, random_state=42, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score
```

```python
linear = LinearRegression()
linear.fit(X_train, y_train)
```

```python
y_ = linear.predict(X_test)
```

```python
auc = roc_auc_score(y_test, y_)
```

```python
auc
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
y_ = rfc.predict(X_test)
```

```python
auc = roc_auc_score(y_test, y_)
```

```python
auc
```

```python
cross_val_score(RandomForestClassifier(max_depth = 4), data_scaler, target,verbose=3).mean()
```

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
```

```python
y_ = xgb.predict(X_test)
```

```python
auc = roc_auc_score(y_test, y_)
auc
```

```python
cross_val_score(XGBClassifier(), data_scaler, target,verbose=3).mean()
```
